```python
import numpy as np
from matplotlib import pyplot as plt

x = np.linspace(-10,10,100)
y = x**2

z = np.ones(len(x))

for i in range(6):
    z[i] = y[i]

k = 7
w = int((k-1)/2)
I = np.ones(7)
for i in range(6,len(x)-3):
    z[i] = (1/(7**2))*np.dot(I,y[i-w:i+w+1])


a = np.ones(len(x))
for i in range(6):
    a[i] = y[i]

for i in range(6,len(x)):
    a[i] = (1/(7**2))*np.dot(I,y[i-k+1:i+1])

plt.plot(x,y)
plt.plot(x,z)
plt.plot(x,a)
plt.show()
```

```python
def gaussian_3d(x, y, mean, std_dev):
    return np.exp(-((x - mean[0])**2 + (y - mean[1])**2  / (2 * std_dev**2)))

x = np.ones((5,5))
y = np.ones((5,5))
mean = [0,0]
std_dev = 1

z = gaussian_3d(x, y, mean, std_dev)
print(z)
```
